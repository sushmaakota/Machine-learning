# Simple Linear Regression - Lab Experiment 2

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values # Except the last column, all the rows copied to X
y = dataset.iloc[:, -1].values  # Only the Last Column values copied to y
```

#  Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0) 

# random_state =0 is used for initializing the internal random number generator, 
# it will decide the splitting of data into train and test 
```

# Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

# Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# # Predicting the result of 5 Years Experience
y_pred = regressor.predict([[5]])
print("The Predicted Salary is =",int(y_pred))
```
